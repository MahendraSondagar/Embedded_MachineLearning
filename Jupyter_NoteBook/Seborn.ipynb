## Visualize Distributions With Seaborn

# Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

```python
# Plotting a Displot

import matplotlib.pyplot as plt
import seaborn as sns

#displot 
sns.displot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
# Plotting a Distplot Without the Histogram

import matplotlib.pyplot as plt
import seaborn as sns

#distplot, histogram = false 
sns.distplot([0,1,4,5,8,6,7], hist= False)
plt.show()
```

```python
# Normal (Gaussian) Distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

# Draw random samples from a normal (Gaussian) distribution.
x = random.normal(0,0.1, 1000)
# print (x)
# display the distplot
sns.distplot(x)
plt.show()
```

```python
# Normal (Gaussian) Distribution with lable & color 

from numpy import random
import matplotlib.pyplot as plt 
import seaborn as sns

# distplot with the random normal 
sns.distplot(random.normal(loc= 0, scale= 0.5, size= 100), label= 'normal', color = 'green')
plt.show()
```

```python
## Binomial Distribution
# Binomial Distribution is a Discrete Distribution.

# It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.

# It has three parameters:

# n - number of trials.

# p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).


# size - The shape of the returned array.

from numpy import random
import matplotlib.pyplot as plt 
import seaborn as sns

x= random.binomial(n =10, p= 0.5, size= 1000)
# print (x)
sns.distplot(x, hist= True, kde= False, color= "orange")
plt.show()
```

```python
# differance between the binomial and normal distribution 

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc= 0, scale= 0.5, size=1000), hist = False, color = "green", label= "Normal")
sns.distplot(random.binomial(n= 2, p = 0.5, size=1000), hist = False, color= "orange", label= "Binomial")
plt.show()
```

```python
# Plotting with the relplot function
# The Seaborn library provides us with relplot() function and this function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. 
# The kind parameter selects the underlying axes-level function to use:

import seaborn as sns
print (sns.get_dataset_names())
df = sns.load_dataset('car_crashes')
# print (df.head)
sns.relplot(data = df, x= 'speeding', y = 'alcohol')
plt.show()
```

```python
# barplo with seaborn

import matplotlib.pyplot as plt
import seaborn as sns

# Display the data-set from the list
print (sns.get_dataset_names())
df = sns.load_dataset('iris')
print (df.head())
sns.barplot(x ='sepal_width', y= 'sepal_length', data = df.head(), label= 'Test')
plt.legend()
plt.show()
```

```python
# count plot suing seaborn 

import matplotlib.pyplot as plt
import seaborn as sns

print (sns.get_dataset_names())
df = sns.load_dataset('titanic')
print (df.head())

#count plot display
sns.countplot(data = df, x= 'survived', hue = 'who')

#tital to graph
plt.title("Titanic")
plt.show()
```

```python
# count plot example

import matplotlib.pyplot as plt
import seaborn as sns

#get the list of the datasets
print (sns.get_dataset_names())

# load the datasets 
df = sns.load_dataset('diamonds')

#display the first few datasets 
print (df.head())

# display the count chart 
sns.countplot(data = df.head(), x= 'table')

plt.title("Diamond table")
plt.show()
```

```python
# Box plot using seaborn 

import matplotlib.pyplot as plt 
import seaborn as sns

print (sns.get_dataset_names())
# loading the dataset 

df = sns.load_dataset('mpg')
print (df.head())

# add the white grits as a style in to the graph
sns.set(style= 'whitegrid')

#display the box plot 
sns.boxplot(x= 'horsepower', y= 'acceleration', data = df.head())
plt.show()
```

```python
# seaborn line plot example

import matplotlib.pyplot as plt 
import seaborn as sns

#get the list of the datasets 
print (sns.get_dataset_names())

#load the dataset 
df = sns.load_dataset('iris')
print (df.head())

#line chart plotting 
sns.lineplot(x= 'sepal_length', y= 'sepal_length', hue = 'species', data = df)
plt.show()
```

```python
# seaborn violineplot 

import matplotlib.pyplot as plt
import seaborn as sns

#get the data set list 
print(sns.get_dataset_names())

#loading the dataset 
df = sns.load_dataset('fmri')

#displaying firt head data 
print(df.head())
sns.set(style='whitegrid')

#plotting the violine graph 
sns.violinplot(x= 'timepoint', y= 'signal', data= df, hue = 'region')

plt.show()
```

```python
# pairplot using seaborn 

import matplotlib.pyplot as plt
import seaborn as sns

#list the datasets available 
print (sns.get_dataset_names())

#load the dataset 
df = sns.load_dataset('iris')
print(df.head())

sns.pairplot(data = df, hue = 'species')
plt.show()
```
